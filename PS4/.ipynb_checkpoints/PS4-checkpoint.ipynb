```python
### Galaxies problem set 4
import numpy as np
import pandas as pd
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
data = "./"
filename = "imgblock4.fits"
with fits.open(data+filename) as f:
    hdr = f[0].header
    disk_img = f[0].data
```

```python
vmin = np.percentile(disk_img, 10)
vmax = np.percentile(disk_img, 99.99)
plt.imshow(disk_img, vmin=vmin, vmax=vmax, cmap="jet")
plt.colorbar()
```

```python
plt.hist(disk_img[50], bins=25)
plt.show()
```

```python
plt.plot(disk_img[50, 50:]) # slice in x-direction / radius
disk_slice = disk_img[50, 80:]
idxs = np.arange(30, 50)
p = np.polyfit(idxs, np.log(disk_slice), 1)
p = np.poly1d(p)
plt.plot(np.exp(p(np.arange(1, 51))))

plt.yscale("log")
```

```python
plt.plot(disk_img[50:, 50]) # slice in y-direction
plt.yscale("log")
```
